```python
# mal hecho (intento mío)

def pal(s: str):
    stack = []
    for c in s:
        stack.append(c)
    for c2 in stack:
        if stack.append(c) == stack.pop(c2)
        return True
    return False
```

```python
def is_palindrome(s:str) -> bool:
    stack = list(s)
    reverse = ''
    while len(stack):
        reverse += stack.pop()
    return s == reverse
```

```python
is_palindrome('alan')
```

```python
is_palindrome('ada')
```

```python
def is_palindrome(s:str) -> bool:
    stack = list(s)
    return s == ''.join([stack.pop() for _ in range(len(stack))])
```
