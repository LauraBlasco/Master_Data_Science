```python
colors = ('lime', 'orange', 'm', 'b', 'pink', 'peru')
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
for color, barrio in zip(colors, req_df['Borough'].value_counts().index):
    req_df[req_df['Borough'] == barrio].plot(kind='scatter', ax=ax, c=color, label=barrio, x='Longitude', y='Latitude', alpha=.5, s=1)
```

```python
req_df.plot(kind='hexbin', x='Longitude', y='Latitude', gridsize=80, figsize=(12,10), colormap='jet', mincnt=1)
```
